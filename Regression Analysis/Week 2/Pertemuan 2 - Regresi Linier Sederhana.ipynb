# Input library

```python
# Import library yang diperlukan  
import numpy as np  
import pandas as pd  
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error, r2_score 
```

# Input data secara manual atau dengan mengimport file .xlsx/.csv

```python
# Membuat DataFrame menggunakan pandas 
sales = [26.52, 12.48, 11.16, 22.2, 15.48, 8.64, 14.16, 15.84, 5.76, 12.72, 10.32, 20.88, 11.04, 11.64, 22.8, 26.88, 15, 29.28, 13.56, 17.52, 21.6, 15, 6.72, 18.6, 11.64]  
youtube = [276.12, 53.4, 20.64, 181.8, 216.96, 10.44, 69, 144.24, 10.32, 239.76, 79.32, 257.64, 28.56, 117, 244.92, 234.48, 81.36, 337.68, 83.04, 176.76, 262.08, 284.88, 15.84, 273.96, 74.76]  
newspaper = [83.04, 54.12, 83.16, 70.2, 70.08, 90, 28.2, 13.92, 1.2, 25.44, 29.04, 4.8, 79.08, 8.64, 55.2, 63.48, 136.8, 66.96, 21.96, 22.92, 64.08, 28.2, 59.52, 31.44, 21.96]  
   
data = pd.DataFrame({  
    'sales': sales,  
    'youtube': youtube,  
    'newspaper': newspaper  
})  

data
```

# Menghitung koefisien secara manual menggunakan rumus regresi

```python
# Menghitung rata-rata dari 'youtube' dan 'sales'  
x_bar = np.mean(data['youtube'])  
y_bar = np.mean(data['sales'])  
  
# Menghitung B1 (slope)  
numerator = np.sum((data['youtube'] - x_bar) * (data['sales'] - y_bar))  
denominator = np.sum((data['youtube'] - x_bar) ** 2)  
B1_manual = numerator / denominator  
  
# Menghitung B0 (intercept)  
B0_manual = y_bar - B1_manual * x_bar  
  
# Menampilkan hasil koefisien secara manual  
print(f"Koefisien B0 (manual): {B0_manual}")  
print(f"Koefisien B1 (manual): {B1_manual}")  
```

# Menggunakan fungsi built-in dari scikit-learn  

```python
# Menyiapkan data untuk model regresi  
X = data[['youtube']]  # Variabel independen (harus berbentuk 2D array)  
y = data['sales']      # Variabel dependen  
  
# Membuat model regresi linear  
model = LinearRegression()  
  
# Melatih model dengan data  
model.fit(X, y)  
  
# Mendapatkan koefisien B0 dan B1 dari model  
B0_sklearn = model.intercept_  
B1_sklearn = model.coef_[0]  

# Menampilkan hasil koefisien menggunakan scikit-learn  
print(f"Koefisien B0 (scikit-learn): {B0_sklearn}")  
print(f"Koefisien B1 (scikit-learn): {B1_sklearn}")  
  
# Melakukan prediksi menggunakan model  
y_pred = model.predict(X)  
  
# Menghitung metrik evaluasi model  
mse = mean_squared_error(y, y_pred)  
r2 = r2_score(y, y_pred)  
  
# Menampilkan metrik evaluasi model  
print(f"Mean Squared Error (MSE): {mse}")  
print(f"R-squared (R2): {r2}") 
```
